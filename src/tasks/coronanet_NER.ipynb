# Named Entity Recognition (NER)

```python
import spacy
from collections import  Counter
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def plot_named_entity_barchart(text):
    nlp = spacy.load("en_core_web_sm")
    
    def _get_ner(text):
        doc=nlp(text)
        return [X.label_ for X in doc.ents]
    
    ent=text.apply(lambda x : _get_ner(x))
    ent=[x for sub in ent for x in sub]
    counter=Counter(ent)
    count=counter.most_common()
    
    # Initialize the matplotlib figure
    f, ax = plt.subplots(figsize=(10, 20))
   
    x,y=map(list,zip(*count))
    sns.barplot(x=y,y=x)
```

```python
plot_named_entity_barchart(df['description'])
```

```python
# 2. Check which countries, places or cities name appear the most in the description

import spacy
from collections import  Counter
import seaborn as sns

def plot_most_common_named_entity_barchart(text, entity="GPE"):
    nlp = spacy.load("en_core_web_sm")
    
    def _get_ner(text,ent):
        doc=nlp(text)
        return [X.text for X in doc.ents if X.label_ == ent]

    entity_filtered=text.apply(lambda x: _get_ner(x,entity))
    entity_filtered=[i for x in entity_filtered for i in x]
    
    # Initialize the matplotlib figure
    f, ax = plt.subplots(figsize=(10, 15))
    
    counter=Counter(entity_filtered)
    x,y=map(list,zip(*counter.most_common(10)))
    sns.barplot(y,x).set_title(entity)
```

```python
plot_most_common_named_entity_barchart(df['description'], entity="GPE")
```

# Sentiment Analysis

The sentiment function of TextBlob returns two properties:

1. Polarity: is a floating-point number that lies in the range of [-1,1] where 1 refers to positive statement and -1 refers to a negative statement.
2. Subjectivity: refers to how someone’s judgment is shaped by personal opinions and feelings. Subjectivity is represented as a floating-point value which lies in the range of [0,1].

```python
from textblob import TextBlob
    
def plot_polarity_histogram(text):
    
    def _polarity(text):
        return TextBlob(text).sentiment.polarity
        
    polarity_score =text.apply(lambda x : _polarity(x))
    polarity_score.hist(figsize=(20,10))
```

```python
plot_polarity_histogram(df['description'])
```

```python
# Classifying the policies as negative, positive and neutral based on the scores

from textblob import TextBlob
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk

def sentiment_vader(text, sid):
    ss = sid.polarity_scores(text)
    ss.pop('compound')
    return max(ss, key=ss.get)

def sentiment_textblob(text):
        x = TextBlob(text).sentiment.polarity
        
        if x<0:
            return 'neg'
        elif x==0:
            return 'neu'
        else:
            return 'pos'

def plot_sentiment_barchart(text, method='TextBlob'):
    if method == 'TextBlob':
        sentiment = text.map(lambda x: sentiment_textblob(x))
    elif method == 'Vader':
        nltk.download('vader_lexicon')
        sid = SentimentIntensityAnalyzer()
        sentiment = text.map(lambda x: sentiment_vader(x, sid=sid))
    else:
        raise ValueError('Textblob or Vader')
    
    plt.bar(sentiment.value_counts().index,
            sentiment.value_counts())
```

```python
plot_sentiment_barchart(df['description'], method='Vader')
```

# Topic modeling exploration with pyLDAvis

```python
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
```

```python
from nltk.corpus import wordnet as wn
from nltk.stem.wordnet import WordNetLemmatizer
```

```python
import spacy    
en = spacy.load("en_core_web_sm")
en.Defaults.stop_words |= {"says", "The", "To", "-", "in", "New", "Is", "A", "new", "Of", "In", "'", "''", "we", "re", "says", "said"}

new_sw_spacy = en.Defaults.stop_words
```

```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')

def preprocess_news(df):
    corpus=[]
    stem=PorterStemmer()
    lem=WordNetLemmatizer()
    for news in df['description']:
        words=[w for w in word_tokenize(news) if (w not in new_sw_spacy)]
        
        words=[lem.lemmatize(w) for w in words if len(w)>2]
        
        corpus.append(words)
    return corpus

corpus=preprocess_news(df)
```

```python
import numpy as np
import nltk
from nltk.corpus import stopwords
import gensim
from nltk.stem import WordNetLemmatizer,PorterStemmer
from nltk.tokenize import word_tokenize
import pyLDAvis.gensim
```

```python
def get_lda_objects(text):
    
    new_sw_spacy = en.Defaults.stop_words

    
    def _preprocess_text(text):
        corpus=[]
        stem=PorterStemmer()
        lem=WordNetLemmatizer()
        for news in text:
            words=[w for w in word_tokenize(news) if (w not in new_sw_spacy)]

            words=[lem.lemmatize(w) for w in words if len(w)>2]

            corpus.append(words)
        return corpus
    
    corpus=_preprocess_text(text)
    
    dic=gensim.corpora.Dictionary(corpus)
    bow_corpus = [dic.doc2bow(doc) for doc in corpus]
    
    lda_model =  gensim.models.LdaMulticore(bow_corpus, 
                                   num_topics = 10, 
                                   id2word = dic,                                    
                                   passes = 10,
                                   workers = 10)
    
    return lda_model, bow_corpus, dic

def plot_lda_vis(lda_model, bow_corpus, dic):
    pyLDAvis.enable_notebook()
    vis = pyLDAvis.gensim.prepare(lda_model, bow_corpus, dic)
    return vis
```

```python
lda_model, bow_corpus, dic = get_lda_objects(df['description'])
```

```python
lda_model.show_topics()
```

```python
plot_lda_vis(lda_model, bow_corpus, dic)
```

On the left side, the area of each circle represents the importance of the topic relative to the corpus. As there are ten topics, we have ten circles.
The distance between the center of the circles indicates the similarity between the topics. Here we can see that the topic 3 and topic 5 overlap, also topic 6 and topic 8. This indicates that the topics are more similar.
On the right side, the histogram of each topic shows the top 30 relevant words. For example, in topic 1 the most relevant words are Canada, health, service, people, public and so on which means it is related to the covid policies.
